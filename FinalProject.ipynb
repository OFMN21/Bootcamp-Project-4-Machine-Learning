# Cleaning Data

```python
# Dropping the unused columns
col = ['App','Current Ver','Android Ver']
df.drop(columns=col, axis=1, inplace=True)
```

# ML

```python
from sklearn import metrics
from sklearn.metrics import confusion_matrix
from sklearn.metrics import accuracy_score
```

```python
# The Type column has two values Paid and Free will change the values to 0 ,1 where 0 is free and 1 is paid.
df['Type'] = df['Type'].map({'Free': 0, 'Paid': 1})
```

```python
X = df[['Type', 'Reviews', 'Installs']]
y = df['Rating']
```

```python
from sklearn.ensemble import RandomForestRegressor, GradientBoostingRegressor
import xgboost as xgb
import lightgbm as lgb
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, train_size=0.75, random_state=0)
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
#Creating a StandardScaler object to normalize the X train and test set feature data
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

# -====================================================================================

```python
gbreg = GradientBoostingRegressor(learning_rate= 0.05, max_depth= 3, n_estimators= 300)
gbreg.fit(X_train, y_train)
y_pred_gbreg = gbreg.predict(X_test)
```

```python
mse_gbreg = mean_squared_error(y_test, y_pred_gbreg)
r2_gbreg = r2_score(y_test, y_pred_gbreg)
```

```python
print('Gradient Boosting Regression:')
print('Mean Squared Error:', mse_gbreg)
print('R-squared:', r2_gbreg)
```
